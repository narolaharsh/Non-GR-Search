```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from pycbc.waveform import get_td_waveform
```

```python
bank = h5py.File(open('attempt10-nonGR.h5', 'r'))
```

```python
mass1 = bank['mass1'][:]
mass2 = bank['mass2'][:]
```

```python
plt.figure(figsize = (8, 6))
plt.scatter(mass1, mass2, s = 1)
plt.title('non-GR Bank in m1-m2 plane')
plt.xlabel('Mass1')
plt.ylabel('Mass2')
plt.xlim(0, 101)
plt.ylim(0, 60)
plt.xticks(np.arange(0, 101, 5))
plt.yticks(np.arange(0, 60, 5))
plt.savefig('non-GR-m1m2.png')
plt.show()
```
